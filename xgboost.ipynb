```python
# for data
import pandas as pd
import numpy as np

# for plotting
import matplotlib.pyplot as plt
import seaborn as sns

# for statistical tests
import scipy
import statsmodels.formula.api as smf
import statsmodels.api as sm

# for machine learning
from sklearn import model_selection, preprocessing, feature_selection
import xgboost
from sklearn.model_selection import StratifiedKFold
from sklearn.model_selection import cross_val_score

from sklearn.preprocessing import MinMaxScaler
```

```python
# label: 
dtf = pd.read_csv('train.csv', header = None)
dtf = dtf.set_index(dtf.columns[0])
# dtf = dtf.set_index("user_id")
dtf.head()
```

```python
dtf = dtf.sample(frac = 1)
X_train = dtf.iloc[:,:-1]
scaler = MinMaxScaler()
X_train = scaler.fit_transform(X_train)

# from sklearn.preprocessing import StandardScaler
# scaler = StandardScaler()
# # Fit on training set only.
# scaler.fit(X_train)
# # Apply transform to both the training set and the test set.
# X_train = scaler.transform(X_train)
# from sklearn.decomposition import PCA
# # Make an instance of the Model
# pca = PCA(.95)
# pca.fit(X_train)
# X_train = pca.transform(X_train)


y_train = dtf.iloc[:,-1:].values.ravel()
print("X_train shape:", X_train.shape, "| y_train shape:", y_train.shape)
```

```python
X_test = pd.read_csv('test.csv', header = None,index_col=0)
X_test.head()
```

```python
# CV model
model = xgboost.XGBClassifier()
model.fit(X_train, y_train)

X_test_scaled = scaler.transform(X_test)
# X_test = scaler.transform(X_test)
# X_test = pca.transform(X_test)
y_predict = model.predict(X_test_scaled)
print(y_predict)
```

```python
Id = [i for i in range (0, 2959)]
predict = pd.DataFrame(Id, columns = ['Id'])
predict['Category'] = y_predict
print(predict)
predict.to_csv('output.csv', index=False)
```
